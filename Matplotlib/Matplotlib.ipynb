```python
#if we increase the number of rows, then figure size get reduce, so that no of rows can get increase
```

```python
plt.subplot(5, 2, 1)  #5 rows, 2 columns
plt.plot(x,y,'m')
plt.subplot(1, 2, 2)  #2 row, 3 columns
plt.plot(x,y2,'c')
```

```python
#Right Now, in above plots points not shown - i.e. which X-point corresponds to which y-point
#The third parameter has three values
# color, shape of point, line shape
# mo   magenta and oval points
# c+-- cyan and plus shape points and dotted line 
```

```python
plt.subplot(1, 2, 1)  #5 rows, 2 columns
plt.plot(x,y,'mo')
plt.subplot(1, 2, 2)  #2 row, 3 columns
plt.plot(x,y2,'c+--')
```

```python
#we can change the markersize as well 
```

```python
plt.subplot(1, 2, 1)  #5 rows, 2 columns
plt.plot(x,y,'mo',markersize=2)  #Now the dots size are very small
plt.subplot(1, 2, 2)  #2 row, 3 columns
plt.plot(x,y2,'c+--')
```

```python
#We have the Object Oriented Way to plot the same grahphs
```

```python
fig = plt.figure()
axes = fig.add_axes([0.1,0.1,1,1])  #i.e. when x is 0.1, y is 0.1 and when x is 1, y is 1

axes.plot(x,y,'g')
axes.set_xlabel('X-axis')
axes.set_ylabel('Y-axis')
axes.set_title('Random Plot using Object Oriented way')
```

```python
#There are so many methods which we can use with axes
```

```python
dir(axes)
```

```python
#The object oriented method is bit complex, but by this way - we have full control
```

```python
# Let's draw two plots in a single figure
```

```python
fig = plt.figure()
ax1 = fig.add_axes([0,0,1,1])
ax2 = fig.add_axes([0.1,0.6,0.4,0.3])

ax1.plot(x,y,'m')
ax1.set_xlabel('x')
ax1.set_ylabel('y')
ax1.set_title('BIG PLOT')

ax2.plot(x,y2,'c')
ax2.set_xlabel('x')
ax2.set_ylabel('y')
ax2.set_title('SMALL PLOT')
```

```python
# So, with object oriented way - we have full control over figures and plots
```

```python
#we have one more function subplots() which is little bit different to subplot()
```

```python
fig,ax = plt.subplots(1,2)

ax[0].plot(x,y,'b')

ax[1].plot(x,y,'r')
```

```python
#another way to plot subplots() using for-loop
```

```python
fig,ax = plt.subplots(1,2)

for axes in ax:
    axes.plot(x,y)    
```

```python
#if we want different colour in both subplots
```

```python
fig,ax = plt.subplots(1,2)
color = ['r','g']
"""
Enumerate() method adds a counter to an iterable and returns it in a form of enumerate object. 7
This enumerate object can then be used directly in for loops.
"""
for i,axes in enumerate(ax):
    axes.plot(x,y,color[i])
```

```python
#We can plot multiple lines in single figure
```

```python
fig,ax = plt.subplots(figsize=(8,4),dpi=100)

ax.plot(x,y,'c',label='xy')
ax.plot(x,y2,'m',label='xy2')

ax.set_xlabel('X-axis')
ax.set_ylabel('Y-axis')
ax.legend()
ax.set_title('BIG SIZE FIGURE')
```

```python
# we can change the location of the legend
```

```python
fig,ax = plt.subplots(figsize=(8,4),dpi=100)

ax.plot(x,y,'c',label='xy')
ax.plot(x,y2,'m',label='xy2')

ax.set_xlabel('X-axis')
ax.set_ylabel('Y-axis')
ax.legend(loc=1)  #0= best, 1 = upper right, 2 = upper left etc etc (we can check other using SHIFT+double tab)
ax.set_title('BIG SIZE FIGURE')
```

```python
#we can also control the ticks (x-ticks and y-ticks)
```

```python
fig,ax = plt.subplots(figsize=(10,4))

ax.plot(x,y2)
ax.set_xticks([1,2,3,5,10])
ax.set_yticks([])  #to remove - we pass empty list
plt.show()
```

```python
#we can set the labels to xticks
```

```python
fig,ax = plt.subplots(figsize=(10,4))

ax.plot(x,y2)
ax.set_xticks([1,3,5,10])
ax.set_xticklabels([r'$\alpha$',r'$\beta$',r'$\gamma$',r'$\delta$'],fontsize=19)
ax.set_yticks([])
plt.show()
```

```python
#we can set the axis-limits in scientific notation as well
```

```python
from matplotlib import ticker

fig,ax = plt.subplots()

ax.plot(x,y2)
ax.set_title('Scientific Notation')

formatter = ticker.ScalarFormatter(useMathText=True)
formatter.set_scientific(True)
formatter.set_powerlimits((-1,-1))

ax.yaxis.set_major_formatter(formatter)
```
