```python
def main():
    N = int(input())
    quantitty_required = [int(i) for i in input().split()]
    quantity_present = [int(i) for i in input().split()]
    out = 999999999
    for i in range(N):
        max = int(quantity_present[i] / quantitty_required[i])
        if max < out:
            out = max
    print(out)
        
main()
```

```python
def main():
 T = int(input())
 for i in range(T):
    win=0
    N = int(input())
    power_g = [int(i) for i in input().split()]
    pow_opponent = [int(i) for i in input().split()]
    power_g.sort(reverse=True)
    pow_opponent.sort(reverse=True)
    c = 0
    for i in range(N):
        for j in range(c,N):
            if power_g[i] > pow_opponent[j]:
                c = j + 1
                win+=1
                break
    print(win)
main()
```
